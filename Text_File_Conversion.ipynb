```python
import os
import seaborn as sns
files=["File.txt","File1.txt","File2.txt","File3.txt","File4.txt"]
a=[200,400,600,800,1000] #File size in MB
# Create files and write content
count=0
for file_path in files:
  f=open(file_path,"wb")
  file_sizee= a[count]*1024*1024 #Converting MB to Bytes
  f.write(b'0' * file_sizee)
  count=count+1
  f.close()
# Modify file contents to uppercase
for File in files:
  F1=open(File,"r")
  content="" #initialized as an empty string ""
  for Line in F1:
    content+=Line.upper() #uppercase line is appended to the content string using the += operator.
  F1.close() 
  f1=open(File,"w")
  for line in File:
    f1.write(content)
  f1.close()
```
